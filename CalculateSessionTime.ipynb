```python
from io import StringIO
df = pd.read_csv(StringIO(table), sep='\s+')
df
```

```python
# define treshold value
#from datetime import timedelta
T = 180000 # 3 mins
```

```python
df
```

```python
df =  pd.concat([df,
                    df.groupby('User').transform(lambda x:x.shift(1))]
                    ,axis=1)

df.columns = ["Time", "Action", 'User', 'prev_Time', 'Action 2']
```

```python
df['new_session'] = ((df['Time'] - df['prev_Time'])>=T).astype(int)
```

```python
# create the session_id
df['increment'] = df.groupby("User")['new_session'].cumsum()
df['session_id'] = df['User'].astype(str) + '_' + df['increment'].astype(str)
```

```python
df2 = t(df).select("User", "Time", "prev_Time", "session_id")
```

```python
df2.head(n=5)
```

```python
df3 = df2.groupby("session_id").agg(
    _min=pd.NamedAgg(column="Time", aggfunc=min),
    _max=pd.NamedAgg(column="Time", aggfunc=max)
)
```

```python
df3 = pd.DataFrame(df3.to_records())
```

```python
df3["_sum"] = df3["_max"] - df3["_min"]
```
